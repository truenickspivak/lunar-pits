```python
import subprocess
from pathlib import Path

import rasterio
import matplotlib.pyplot as plt
import numpy as np


PROJECT = Path(r"C:\Users\nicks\desktop\lunar-pits")
SCRIPT = PROJECT / "scripts" / "process_lroc_product.py"
OUT_DIR = PROJECT / "data" / "processed" / "lroc_tif"


def process_lroc(product_id, pixel_resolution=0.5, keep_temp=False):
    product_id = product_id.upper().replace(".IMG", "").replace(".XML", "")

    cmd = [
        "python",
        str(SCRIPT),
        product_id,
        "--pixel-resolution",
        str(pixel_resolution),
    ]

    if keep_temp:
        cmd.append("--keep-temp")

    print("Running:")
    print(" ".join(cmd))

    subprocess.run(
        cmd,
        cwd=PROJECT,
        check=True,
    )

    tif_path = OUT_DIR / f"{product_id}.map.tif"

    if not tif_path.exists():
        raise FileNotFoundError(f"Expected output not found: {tif_path}")

    print("Output TIFF:")
    print(tif_path)

    return tif_path


def show_lroc_tif(tif_path):
    tif_path = Path(tif_path)

    with rasterio.open(tif_path) as src:
        print("Width:", src.width)
        print("Height:", src.height)
        print("Bands:", src.count)
        print("Data type:", src.dtypes[0])
        print("Bounds:", src.bounds)
        print("CRS:", src.crs)
        print("Transform:", src.transform)

        img = src.read(1)

    print("Image shape:", img.shape)
    print("Pixel value range:", np.nanmin(img), "to", np.nanmax(img))

    plt.figure(figsize=(12, 12))
    plt.imshow(img, cmap="gray")
    plt.axis("off")
    plt.show()

    return img
```

```python
product_id = "M114328462RE"

tif_path = process_lroc(product_id, pixel_resolution=0.5)
img = show_lroc_tif(tif_path)
```
