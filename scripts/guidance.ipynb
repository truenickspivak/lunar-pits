```python
import sys
import torch
import torchvision
import rasterio
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Python:", sys.version.split()[0])
print("PyTorch:", torch.__version__)
print("torchvision:", torchvision.__version__)
print("rasterio:", rasterio.__version__)
print("numpy:", np.__version__)
print()
print("CUDA available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
    print("CUDA version:", torch.version.cuda)
else:
    print("Running on CPU only")
```

```python
from torchvision.models.detection import (
    maskrcnn_resnet50_fpn_v2,
    MaskRCNN_ResNet50_FPN_V2_Weights
)

print("Loading pretrained Mask R-CNN model...")
model = maskrcnn_resnet50_fpn_v2(weights=MaskRCNN_ResNet50_FPN_V2_Weights.DEFAULT)

model.eval()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print(f"Model loaded on: {device}")

dummy_image = torch.randn(3, 800, 800).to(device)
with torch.no_grad():
    predictions = model([dummy_image])

print(f"\nNumber of detections: {len(predictions[0]['boxes'])}")
print(f"Output keys: {list(predictions[0].keys())}")
print(f"Boxes shape: {predictions[0]['boxes'].shape}")
print(f"Masks shape: {predictions[0]['masks'].shape}")
print(f"Scores shape: {predictions[0]['scores'].shape}")
print("\nForward pass succeeded.")
```
